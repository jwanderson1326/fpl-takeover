```python
from fplpandas import FPLPandas
import numpy as np
import pandas as pd
```

```python
fpl = FPLPandas()
```

```python
teams = fpl.get_teams()
teams
```

```python
#player_summary, player_hist, player_fixtures, player_future = fpl.get_players()
```

```python
player_dfs = fpl.get_players()
```

```python
fixtures = fpl.get_fixtures()
```

```python
#Clean the Player Summary Table
player_summary = player_dfs[0]
convert_dict = {'value_season': float} 
player_summary = player_summary.astype(convert_dict)
player_summary['team'] = player_summary.team_code.map(teams.set_index('code').name)
```

```python
#Clean the Player Fixtures Table
player_fixtures = player_dfs[2]
convert_dict_fixture = {
    'element': int
}

player_lookup = player_summary[['web_name', 'element_type', 'team', 'minutes']]
team_lookup = teams[['name','strength']]
player_fixtures = player_fixtures.astype(convert_dict_fixture)

player_fixtures = pd.merge(player_fixtures, player_lookup, left_on='element', right_on='id', how='left')
player_fixtures = pd.merge(player_fixtures, team_lookup, left_on='opponent_team', right_on='id', how='left')
player_fixtures = player_fixtures[player_fixtures['minutes_y']>=0]
```

```python
fixtures.to_csv('fixtures.csv')
```

```python
player_summary[['web_name','ep_this','ep_next',"element_type"]].sort_values(by=["ep_next"], ascending=False).to_csv("adhoc_ep_this.csv")
```
